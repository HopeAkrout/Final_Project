## Merge Gun Data, Gun Laws, Gun Violence

### Merge Gun Laws and Gun Violence

```python
gun_violence_2016 = gun_violence_2016.groupby(['state']).sum()
gun_violence_2016 = gun_violence_2016.drop(['year'], axis=1)
gun_violence_2016 
```

```python
merged_violence_laws = pd.merge(gun_violence_2016, gun_laws_2016, how='inner', left_on=['state'], right_on=['state'])
merged_violence_laws.head()
```

```python
merged_violence_laws.set_index('state')
```

```python
merged_violence_laws_transposed = merged_violence_laws.transpose()
```

```python
merged_violence_laws_transposed = merged_violence_laws_transposed.rename(columns=merged_violence_laws_transposed.iloc[0]).iloc[1:]
```

### Merge Gun Data and Gun Violence

```python
merged_gun_data_violence = pd.merge(cleaned_gun_data_2016, cleaned_gun_violence_2016, how='inner', left_on=['state'], right_on=['state'])
merged_gun_data_violence
```

```python
merged_gun_data_violence.set_index('state')
```

```python
merged_gun_data_violence_transposed = merged_gun_data_violence.transpose()
merged_gun_data_violence_transposed
```

```python
merged_gun_data_violence_transposed = merged_gun_data_violence_transposed.rename(columns=merged_violence_laws_transposed.iloc[0]).iloc[1:]
```
