## Merge all back into one dataset

```python
# setting the path for joining multiple files
files = os.path.join('/content/Drive/MyDrive/Final_Project/segment_2_machine_learning/Cleaning_ML/to_merge/merged*.csv')

# list of merged files returned
files = glob.glob(files)

# joining files with concat and read_csv
merged_df = pd.concat(map(pd.read_csv, files), ignore_index=True)
```

```python
merged_df.head()
```

```python
merged_df.tail()
```

```python
# merge state and year into one column, replace as index and drop 'year' and 'state' columns
merged_df["state_year"] = merged_df['state'] +"_"+ merged_df['year_x'].astype(str)
merged_df.drop(['year_x','year_y','state'], axis=1, inplace=True)
merged_df.sample(5)
```

```python
# find data
merged_df.deaths.describe()
```

```python
# splitting death rate by above and below national mean

merged_df['deaths_above_below_mean'] = np.where(deaths_df['deaths'] < 795, 'below_national_mean', 'above_national_mean')
merged_df.head(5)
```

```python
merged_df = merged_df.reindex(columns=['state_year', 'lawtotal', 'rate', 'deaths', 'deaths_above_below_mean'])
merged_df
```
